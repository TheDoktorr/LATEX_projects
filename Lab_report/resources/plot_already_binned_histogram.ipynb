### Plotting a histogram if you already have binned values

In week 17 we saw how to create a histogram using the function `hist()` from the module `matplotlib.pyplot`. This function bins a list of given values and then plots the result as a histogram. Unless otherwise specified, it also tries to determine a sensible range and size of bins automatically.

If you already have binned data, for example counts of events, then the `hist()` function is not suitable. Instead, you should use the function `bar()` and send as inputs the values corresponding to the centre of the bins as a list and the number of items in each bin as a second list. The cell below shows an example of using `bar()`.

```python
x_data = [0, 1, 2, 3, 4, 5, 6, 7]
y_data = [1, 2, 3, 5, 7, 3, 3, 1]

plt.figure(figsize=(6,4))

plt.bar(x_data, y_data)

plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
