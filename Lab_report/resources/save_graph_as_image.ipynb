```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Load some data to plot

```python
x_values = np.arange(1, 8, 1)
y_data = [x**2 + (2*np.random.normal()) for x in x_values]  # This is not real data,
y_fit = [x**2 for x in x_values]  # ... and this is not a real fit to the data!
```

### Saving pyplot figures as images

A figure made using `matplotlib.pyplot` can be saved to an image file using the function `savefig()`. This function **must be used before** `show()`. If it is called afterwards, the image file will be empty.

The input to `savefig()` is the name of the file that the image will be saved to. The format of the image is determined by the *file extension* (characters at the end of the filename, after a dot). Image formats that are supported include PDF, JPG and PNG (specified by `.pdf`, `.jpg` or `.png` at the end of the file name.)

Before using `savefig()`, it is often helpful to call the function `tight_layout()`. This function rearranges the elements of your plot so that they fit cleanly within the figure. If you do not do this, parts of your plot can sometimes end up missing from the final images.

In the example below, a graph is plotted and saved as both a PDF and a JPG image. After running the code, you can go to the Jupyter "Home Page" tab in your web browser to open the two files and see what they look like. Look for the files `test_figure.pdf` and `test_figure.jpg`. Zoom in on both images - which ones looks better?

```python
plt.figure(figsize=(7,4))

plt.errorbar(x_values, y_data, yerr=3, fmt='ob', capsize=4, label='Data')
plt.plot(x_values, y_fit, '--k', label='$x^2$')

plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.grid()

plt.tight_layout()
plt.savefig('test_figure.pdf')
plt.savefig('test_figure.jpg')

plt.show()
```
